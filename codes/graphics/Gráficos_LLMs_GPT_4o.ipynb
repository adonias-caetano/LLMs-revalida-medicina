**Objective:** Gera os gráficos do GPT-4o versus LLama 3.1

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = '<public link to access the spreadsheet of number of correct answers by LLM>'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id

dados = pd.read_excel(read_url, sheet_name='En-Accuracy_GPT-4o_LLama')
dados
```

```python
# Configurando o tema e estilo do gráfico
sns.set_theme(rc={'figure.figsize': (6, 7)})
sns.set(font_scale=1.5)
sns.set_style('white')

# Definindo uma nova paleta de cores
paleta_cores = sns.color_palette('viridis', len(dados))

fig, ax = plt.subplots()

# Criando o gráfico de barras
ax = sns.barplot(data=dados, x="Model", y="Number of correct answers", hue="Approach", palette=paleta_cores)

#plt.xticks([])
plt.ylabel('Number of correct answers')
plt.yticks([0, 25, 50, 75, 100])

# Adicionando rótulos nas barras
for container in ax.containers:
    bars = container
    ax.bar_label(bars, labels=[f'{x.get_height():.0f}'.replace('.', ',') for x in bars],
                 fontsize=16, rotation=0, padding=3)

plt.show()

```
