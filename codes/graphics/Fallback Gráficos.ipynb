```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Gemini e Copilot**

```python
url = 'public link to access the Fallbacks-LLMs spreadsheet'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id
```

```python
dados_all = pd.read_excel(read_url, sheet_name='Fallback_Gemini_Copilot')
dados_all['Taxa de Acerto'] = dados_all['Taxa de Acerto']* 100
dados_all
```

```python
sns.set_theme(rc={'figure.figsize':(20, 10)})
#increase font size of all elements
sns.set(font_scale=1.5)
ax = sns.barplot(dados_all, x="Edição do Revalida", y="Taxa de Acerto", hue="Tipo de Entrada", palette = ['blue', 'red', '#99CCFF', '#FF9999'])
plt.legend(title='')

plt.ylabel('Fallback Index')
plt.yticks([0, 2, 4, 6, 8, 10])


plt.title(
    "Fallback",
    fontsize='large',
    loc='left',
    fontweight='bold',
    style='italic',
    family='monospace')

for container in ax.containers:
  bars = container
  ax.bar_label(bars, labels = [f'{x.get_height():.2f}%'.replace('.', ',') if x.get_height() > 0 else f'{x.get_height():.0f}%'.replace('.', ',') for x in bars], fontsize=16, rotation=90, padding=3)

plt.show()
```
