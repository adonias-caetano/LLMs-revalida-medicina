```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
paleta_cores = ['yellow', 'purple', 'orange', 'darkviolet']
```

```python
def show_grafico(dados, name_model, y_name, size, yticks):
  sns.set_theme(rc={'figure.figsize':size})
  #increase font size of all elements
  sns.set(font_scale=1.5)
  ax = sns.barplot(dados, x="Revalida Edition", y=y_name, hue="Model", palette = paleta_cores)
  plt.legend(title='')

  sns.set_style(style='white')

  plt.title(
      name_model,
      fontsize='large',
      loc='left',
      fontweight='bold',
      style='italic',
      family='monospace')

  plt.yticks(yticks)

  for container in ax.containers:
    bars = container
    ax.bar_label(bars, labels = [f'{x.get_height():.2f}%' for x in bars], fontsize=16, rotation=90, padding=3)

  plt.show()
```

```python
url = 'public link to access the Fallbacks-LLMs spreadsheet'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id
```

## **Accuracy Rate**



```python
dados = pd.read_excel(read_url, sheet_name='Accuracy')
dados['Accuracy'] = dados['Accuracy']* 100
dados
```

```python
name_model = ' '
show_grafico(dados, name_model, "Accuracy", (20, 5), [0, 25, 50, 75, 100])
```

## **Fallback Index**



```python
dados = pd.read_excel(read_url, sheet_name='Fallback')
dados['Fallback Index'] = dados['Fallback Index']* 100
dados
```

```python
name_model = ' '
show_grafico(dados, name_model, "Fallback Index", (20, 10), [0, 25, 50, 75, 100, 125, 150])
```
