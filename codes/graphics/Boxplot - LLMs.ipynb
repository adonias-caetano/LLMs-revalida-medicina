**Objective:** Gera os boxplots dos sLLMs (chatgpt, gemini e copilot)

```r
# Criando o data.frame com nomes de colunas
copilot <- data.frame(Valor = c(54.43, 70.88, 52.17, 55.68, 54.43, 61.62, 61.72, 61.62, 79.54, 59.49, 83.54, 67.04, 64.13, 57.64, 60.46, 70.37, 58.42, 76.13))
chatgpt <- data.frame(Valor = c(54.43, 66.66, 62.50, 56.32, 68.60, 67.07, 61.79, 55.68, 59.49, 45.56, 65.59, 67.04, 51.72, 68.60, 71.95, 53.93, 53.40))
gemini <- data.frame(Valor = c(64.55, 56.96, 51.08, 52.17, 56.47, 56.97, 66.66, 67.41, 59.09, 56.96, 62.02, 52.17, 60.86, 60, 56.97, 67.90, 62.92, 52.27))
```

```r
xtick<-seq(45, 85, by=5)
```

```r
get_labels <- function(valores) {

  lista <- c() # Inicializa uma lista vazia
  for (numero in valores) {
    str_valor <- toString( numero )
    lista <- c(lista, gsub("\\.", ",", str_valor))
  }
  return (lista)
}
```

```r
# Criando o boxplot
boxplot(copilot$Valor, xlab = "Copilot/GPT-4", col = rgb(0, 1, 1, alpha = 0.25), border = "black")
axis(2, at = xtick, labels = xtick)
text(y = fivenum(copilot$Valor), labels = get_labels( fivenum(copilot$Valor) ), x=0.75)
```

```r
# Criando o boxplot
boxplot(gemini$Valor, xlab = "Gemini", col = rgb(0, 1, 1, alpha = 0.25), border = "black")
axis(2, at = xtick, labels = xtick)
text(y = fivenum(gemini$Valor), labels = get_labels( fivenum(gemini$Valor) ), x=0.75)
```

```r
# Criando o boxplot
boxplot(chatgpt$Valor, xlab = "ChatGPT/GPT-3.5",  col = rgb(0, 1, 1, alpha = 0.25), border = "black")
axis(2, at = xtick)
text(y = fivenum(chatgpt$Valor), labels = get_labels( fivenum(chatgpt$Valor) ), x=0.75)
```
