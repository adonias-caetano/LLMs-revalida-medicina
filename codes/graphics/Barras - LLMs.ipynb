```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
paleta_cores = ['blue', 'red', 'green']
```

## **Copilot**



```python
url = 'public link to access the Fallbacks-LLMs spreadsheet'
file_id = url.split('/')[-2]
read_url='https://drive.google.com/uc?id=' + file_id
```

```python
dados = pd.read_excel(read_url, sheet_name='Copilot')
dados['Taxa de Acerto'] = dados['Taxa de Acerto']* 100
dados
```

```python
sns.set_theme(rc={'figure.figsize':(20, 14)})
#increase font size of all elements
sns.set(font_scale=1.5)
ax = sns.barplot(dados, x="Edição do Revalida", y="Taxa de Acerto", hue="Tipo de Entrada", palette = paleta_cores)
plt.legend(title='')

sns.set_style(style='white')

plt.title(
    "Copilot/GPT-4",
    fontsize='large',
    loc='left',
    fontweight='bold',
    style='italic',
    family='monospace')

plt.yticks([0, 25, 50, 75, 100])



for container in ax.containers:
  bars = container
  ax.bar_label(bars, labels = [f'{x.get_height():.2f}%'.replace('.', ',') for x in bars], fontsize=16, rotation=90, padding=3)

plt.show()
```
