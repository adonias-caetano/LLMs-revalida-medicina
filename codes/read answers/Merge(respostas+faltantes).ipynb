```python

df_respostas1 = pd.read_excel("resposta_gabarito.xlsx")
df_respostas2 = pd.read_excel("faltantes.xlsx")


# Ler o arquivo e imprimir cada linha
respostas = {}
respostas['questao'] = np.linspace(1, 100, 100, dtype=int)

for sheet_name in sheet_names:

  print(sheet_name)

  lista_respostas1 = df_respostas1['respostas_'+sheet_name].to_list()
  lista_respostas2 = df_respostas2['Repostas-'+sheet_name].to_list()
  lista_gabaritos = gabaritos[sheet_name]['GABARITO'].to_list()

  name_resposta = f"respostas_{sheet_name}"
  name_gabarito = f"gabarito_{sheet_name}"

  respostas[name_resposta] = []
  respostas[name_gabarito] = []
  questao = 1

  print("Exame " + sheet_name)

  for resposta1, resposta2, gabarito in zip(lista_respostas1, lista_respostas2, lista_gabaritos):
    if resposta1 == resposta2:
      respostas[name_resposta].append(resposta1)
      respostas[name_gabarito].append(gabarito)
    elif resposta1 == gabarito:
      respostas[name_resposta].append(resposta1)
      respostas[name_gabarito].append(gabarito)
    elif resposta1 != "VOID":
      respostas[name_resposta].append(resposta1)
      respostas[name_gabarito].append(gabarito)
    else:
      respostas[name_resposta].append(resposta2)
      respostas[name_gabarito].append(gabarito)

    questao += 1
```

```python
df = pd.DataFrame(respostas)
df
```

```python
df.to_excel('resposta_gabarito.xlsx', index=False)
```
