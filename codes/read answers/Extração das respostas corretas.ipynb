```python
import pandas as pd
import re
import random
import numpy as np
from google.colab import files
```

## **Loading data**

```python
#sheet_names = ["2017.01", "2017.02", "2020.01", "2020.02"]
sheet_names = ["2017.01", "2017.02", "2020.01", "2020.02", "2021.01", "2022.01", "2022.02", "2023.01", "2023.02"]
```

```python

# Inicializando variáveis
respostas = {}

file_name = f"resposta_gabarito.xlsx"
df_respostas = pd.read_excel(file_name)

#Inicializa
respostas['perguntas'] = []
respostas['respostas'] = []

for sheet_name in sheet_names:

    #Ler o enunciado das perguntas
    file_name = f"respostas-{sheet_name}.xlsx"
    df_perguntas = pd.read_excel(file_name)
    lista_perguntas = df_perguntas[f'Perguntas-{sheet_name}'].to_list()

    #Ler o gabarito e a resposta das perguntas
    lista_respostas = df_respostas[f"respostas_{sheet_name}"].to_list()
    lista_gabaritos = df_respostas[f"gabarito_{sheet_name}"].to_list()
    lista_explicacoes = df_perguntas[f"Repostas-{sheet_name}"].to_list()

    print(f"Exame {sheet_name}")

    for pergunta, resposta, gabarito, explicacao in zip(lista_perguntas, lista_respostas, lista_gabaritos, lista_explicacoes):

        if gabarito not in ["IMAGEM", "TABELA", "NULA"] and resposta == gabarito:
          respostas['perguntas'].append(pergunta)
          respostas['respostas'].append(explicacao)

print("Processamento concluído!")
df = pd.DataFrame(respostas)
df.to_excel('respostas_corretas.xlsx', index=False)
files.download('respostas_corretas.xlsx')
```
